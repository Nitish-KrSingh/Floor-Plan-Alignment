```python
import cv2
import numpy as np

# Load the image
image_path = r"C:\Users\nitis\python_project\imageRotation\img3.jpg"
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise (optional, can improve edge detection)
gray_blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detect edges using Canny edge detector
edges = cv2.Canny(gray_blurred, 50, 150, apertureSize=3)

# Detect lines using Hough Line Transform
lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)

# Create a copy of the original image for drawing
image_with_lines = image.copy()

# Check if any lines are detected
if lines is not None:
    # Draw the lines on the image
    for rho, theta in lines[:, 0]:
        # Convert theta from radians to degrees
        angle = (theta * 180 / np.pi) - 90
        if angle < -45:
            angle += 90
        elif angle > 45:
            angle -= 90

        # Compute the endpoints of the lines
        a = np.cos(theta)
        b = np.sin(theta)
        x0 = a * rho
        y0 = b * rho
        x1 = int(x0 + 1000 * (-b))
        y1 = int(y0 + 1000 * (a))
        x2 = int(x0 - 1000 * (-b))
        y2 = int(y0 - 1000 * (a))

        # Draw the line
        cv2.line(image_with_lines, (x1, y1), (x2, y2), (0, 0, 255), 2)

    # Display the image with lines
    cv2.imshow('Detected Lines', image_with_lines)
    cv2.waitKey(0)
    cv2.destroyAllWindows()

    # Compute the angle of the lines
    angles = []
    for rho, theta in lines[:, 0]:
        angle = (theta * 180 / np.pi) - 90
        if angle < -45:
            angle += 90
        elif angle > 45:
            angle -= 90
        angles.append(angle)

    # Compute the median angle
    median_angle = np.median(angles)
    print(f"Median Angle: {median_angle} degrees")

else:
    raise ValueError("No lines detected in the image.")
```

```python
import matplotlib.pyplot as plt

# Convert image to RGB for displaying with matplotlib
image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

# Display the original image
plt.figure(figsize=(10, 10))
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(image_rgb)
plt.axis('off')

# Display the image with detected lines
image_with_lines_rgb = cv2.cvtColor(image_with_lines, cv2.COLOR_BGR2RGB)
plt.subplot(1, 2, 2)
plt.title('Image with Detected Lines')
plt.imshow(image_with_lines_rgb)
plt.axis('off')

plt.show()
```

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Load the image
image_path = r"C:\Users\nitis\python_project\imageRotation\img3.jpg"
image = cv2.imread(image_path)

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Apply GaussianBlur to reduce noise
gray_blurred = cv2.GaussianBlur(gray, (5, 5), 0)

# Detect edges using Canny edge detector
edges = cv2.Canny(gray_blurred, 50, 150, apertureSize=3)

# Detect lines using Hough Line Transform
lines = cv2.HoughLines(edges, 1, np.pi / 180, 200)

# Create a copy of the original image for drawing
image_with_lines = image.copy()

# Check if any lines are detected
if lines is not None:
    angles = []
    for rho, theta in lines[:, 0]:
        # Convert theta from radians to degrees
        angle = (theta * 180 / np.pi) - 90
        if angle < -45:
            angle += 90
        elif angle > 45:
            angle -= 90
        angles.append(angle)

    # Compute the median angle
    median_angle = np.median(angles)
    print(f"Median Angle: {median_angle} degrees")

    # Correct the rotation direction based on the median angle
    # We need to rotate the image to make the lines horizontal, so we use the negative of the median angle
    rotation_angle = +median_angle  # We rotate in the opposite direction to correct the tilt

else:
    raise ValueError("No lines detected in the image.")

# Get image dimensions
(h, w) = image.shape[:2]
center = (w // 2, h // 2)

# Compute the rotation matrix
M = cv2.getRotationMatrix2D(center, rotation_angle, 1.0)

# Calculate new bounding box dimensions to avoid cropping
cos = np.abs(M[0, 0])
sin = np.abs(M[0, 1])
new_w = int((h * sin) + (w * cos))
new_h = int((h * cos) + (w * sin))

# Adjust the rotation matrix to center the image
M[0, 2] += (new_w / 2) - center[0]
M[1, 2] += (new_h / 2) - center[1]

# Perform the affine transformation
rotated_image = cv2.warpAffine(image, M, (new_w, new_h), flags=cv2.INTER_CUBIC)

# Convert rotated image to RGB for displaying with matplotlib
rotated_image_rgb = cv2.cvtColor(rotated_image, cv2.COLOR_BGR2RGB)

# Display the original image and the rotated image
plt.figure(figsize=(12, 6))

# Original Image
plt.subplot(1, 2, 1)
plt.title('Original Image')
plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
plt.axis('off')

# Rotated Image
plt.subplot(1, 2, 2)
plt.title(f'Rotated Image (Angle: {rotation_angle:.2f} degrees)')
plt.imshow(rotated_image_rgb)
plt.axis('off')

plt.show()

# Save the rotated image
cv2.imwrite('rotated_floor_plan_image_corrected.jpg', rotated_image)
```
